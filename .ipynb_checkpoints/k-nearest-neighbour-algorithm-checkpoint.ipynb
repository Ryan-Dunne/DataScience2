### k-Nearest Neighbor Algorithm for a toy problem
**Example:** classify the type of drug a patient should be prescribed
* The training set consists of only four patients 1-4 with two features: `Na/K ratio, age`, and one target variable: `drug`  
* Our task is to classify the type of drug   a new patient 0 should be prescribed, for example, a patient is 25-years-old and has a Na/K ratio of 13


```python
import pandas as pd 
df = pd.read_csv('/kaggle/input/aimldata/drug-4instances.csv') #Read data to a data frame
df
#df.to_latex("drug-4instances.txt")
```

### Instance-based learning
k-Nearest Neighbour algorithm belongs to supervised learning and is a kind of **instance-based** learning

* First, we have a set of instances (Patients 1 to 4)
* Next, for the new sample (Patient 0), we choose $k$ instances most similar to the new sample. $k$ is a parameter
* Then, we assign the class label (drug) to the new sample based on its neighbour instances' label. This depends on the value of $k$ 

k-Nearest Neighbour is used most often for classification, and also applicable to estimation and prediction tasks

### k-Nearest Neighbor Algorithm: k=1
* Scatter plot of `Na/K` against `Age` shows the patients. Patient 0 with unknown label is the blue point.
* Blue point is closest to (or most similar to) green point (similar `age` and similar `Na/K` ratio)
* Thus Green point is taken to vote
* Because Green point is with label `AX`, this drug is assigned  to blue point
* Here we compare a new sample with only one instance. This is called 1-nearest neighbour (k=1)

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig = plt.figure()
sns.scatterplot(x = "Age", y = "Na/K", hue  = 'Drug', data = df,s=200) #s=200: set marker sizes =200
plt.show() 
```

###  k-Nearest Neighbor Algorithm: k=3

* Compare the new record with 3 instances. 
* This is called 3-nearest neighbour.
* Each instance ( (one Green, one Red, one Orange point) in the neighbour of new patient (blue point) can be taken to vote drug equally
* Tie: the probability of assigning drug `AX, Y, BC` is equal = 1/3. 
* In this case, the simple voting scheme does not help 
* Better strategy: Think about weighting on distance

### k-Nearest Neighbor Algorithm: k=4

* Compare with 4 instances: k=4 
* The 4-nearest neighbour of blue point include 2 red points and 2 other colours
* Based on voting,  drugs `BC` is chosen
* Distance is not considered. Voting weights are uniformly distributed over voters
* Distance is an important factor. Green is closest, so `AX` is more likely to be chosen.

### Issues in k-Nearest Neighbor Algorithm 

* How many neighbors should be used? k = ?
* How is the distance between two points measured?
* How is the information from neighbors combined when making a classification decision (voting)?
* Should all points be weighted equally, or should some points have more influence?


### Choosing k
* What value of k is optimal?
* There is no an obvious answer

**Smaller k**
* Choosing a small value for k may lead the algorithm to overfit the data
* For example, $k=1$, the classification outcome is greatly decided by the the closest instances (perhaps an outlier or exception patient)

**Larger k**
* Larger values will tend to smooth  data values in the training set
* It the values become too large (for example, all training instances), locally interesting values will be overlooked

Choosing the appropriate value for k requires balancing these considerations

## Part 2 Distance or similarity
### Distance function
The similarity between two instances is measure by a distance function
* **distance metric** is a real-valued function $d(x,y)$ used to measure the similarity between data points $x,y$ with properties:
* Property 1: Distance is always non-negative: $d(x,y) \ge 0$
* Property 2: Commutative, distance from $x$ to $y$ is distance from $y$ to $x$
* Property 3: Triangle inequality holds, distance from $x$ to $z$ must be less than or equal to distance from $x$ to $y$ to $z$

 


### Euclidean Distance
Euclidean Distance  is commonly-used to measure distance
* Let  $x=(x_1, \cdots, x_n)$ and $y=(y_1, \cdots, y_n)$, i.e., two data points with $n$ numerical features
* $d(x, y)= \sqrt{ (x_1-y_1)^2 + \cdots (x_n-y_n)^2)}$

Example
* Patient $x$ is 20-years-old and has a Na/K ratio = 12, i.e., $x=(20,12)$
* Patient $y$ is 30-years-old and has a Na/K ratio = 8, i.e., $y=(30,8)$
* Euclidean distance between  instances $x, y$
$$d(x, y)= \sqrt{ (x_1-y_1)^2 + \cdots (x_n-y_n)^2)} = \sqrt{(30-20)^2+(12-8)^2} =10.17$$

### Manhattan Distance
Manhattan Distance  is commonly-used to measure distance

* Let  $x=(x_1, \cdots, x_n)$ and $y=(y_1, \cdots, y_n)$, i.e., two data points with $n$ features
* $d(x, y)= |x_1-y_1| + \cdots + |x_n-y_n|$

Example 
* Patient $x$ is 20-years-old and has a Na/K ratio = 12, i.e., $x=(20,12)$
* Patient $y$ is 30-years-old and has a Na/K ratio = 8, i.e., $y=(30,8)$
* Manhattan distance between these instances 
$$d(x, y)= |x_1-y_1| + \cdots +|x_n-y_n| = |30-20|+|12-8| =14$$
* The choice of a distance function will influce the similarity between two data points, and then the classification results

### Distance metric in Sklearn
`sklearn.neighbors.DistanceMetric` provide different distance functions, called `metric`

* `KNeighborsClassifier(metric=“euclidean”)`

* `KNeighborsClassifier(metric=“manhattan”)`


## Part 3 Data Preparation in KNN
### Data normalization: numerical features 
* When measuring the distance, one or more attributes can have very large values, relative to the other attributes

Example
* `income` range is  30,000 - 100,000
* `years_of_service` range is 0-30

In this case, the values of `income` will overwhelm the contribution of `years_of_service` in distance calculation

To avoid this situation, we will need data  normalization
 

* Min-Max Scaler
    * `income`:     $         x' = \frac{x-30000}{100000-30000}    $ 
    * `years of service`:         $x'= \frac{x-0}{30-0}$
* Standard Scaler

### Data transformation: Categorical features 

* The distance function can not be applied directly to categorical values.

* How to measure the distance between two modules "maths", "physics" or "English", "chemistry"?
* Need data transform from a categorical feature to numerical one
* We may apply Ordinal Encoder to encode each categorical value to a unique a number

"maths = 1", "physics = 2", "English = 3", "chemistry = 4"

* But this is not a good solution if the data is not ordinal. What is the meaning of Manhattan distance(English, chemistry) = 1 < distance (chemistry, maths) = 3?
* Can we find a better solution? 




### OneHot Encoder
* OneHot Encode is used to measure the “difference” with 0, 1

"maths = 0001"

"physics = 0010", 

"English = 0100"

"chemistry = 1000"


### Example: data normalization and transformation
* Let Patient $x$ = 50-year-old male
* Patient $y$ = 20-year-old male
* Patient $z$ = 50-year-old female
* Suppose that Age variable has a range = 50, i.e, minimum = 10, maximum=60
* MinMax normalize age to

`x_Age = 0.8
y_Age =0.2
z_Age = 0.8`

* Encode `Male = 0, 1`, Female = `1, 0`
* Calculate the distance
$$d(x,y) =\sqrt{(0.8-0.2)^2+(0-0)^2+(1-1)^2} =0.6$$

$$d(x,z) =\sqrt{(0.8-0.8)^2+(1-0)^2+(0-1)^2} =1.4$$

*In this case, Patient $x$ is now closer to Patient $y$ than to Patient $z$

## Part 4 Combination Function
## Combination Function

* Given a value of K, a distance function is used to determine which k instances are most similar to the new  record
* How do the k similar instances combine to provide a classification decision for the new record?
* Need a combination function. 
* Example: new patient’s neighbour consists of 3 similar instances
* How to vote?



### Simple Unweighted Voting
* First, decide  how many instances  can have a vote in classifying the new record. k=3
* Then, find the 3 nearest neighbour instances to the new data, measured by the distance  
* Once the 3 instances are chosen, then taking simple unweighted voting; each instance have one vote, regardless of the distance 

* However, simple unweighted voting did not help and resulted in a tie 
* Perhaps weighted voting should be considered?


### Weighted Voting
* Intuitively, an instance that are closer or more similar to the new record should have a larger weight on voting than an instance that is less similar
* In weighted voting, the vote weight of an instance  is inversely proportional to the distance between a  neighbor instance $x$ and new record $x_{new}$
    $$
    weight =\frac{1}{distance(x,x_{new})}
    $$


### Voting  in Sklearn
`sklearn.neighbors.KNeighborsClassifier (weights=’uniform’)`

`weights` : default = ‘uniform’

1. **‘uniform’** : uniform weights (Simple Unweighted Voting). All instances in neighbourhood are weighted equally.
2. **‘distance’** : weight points by the inverse of their distance (Weighted Voting). The closer, the more simlar.



## Part 5 Case Study: KNN for predicting credit risk
### Data and Task: Predict Credit Risk
* Predict credit `risk` based on `mortgage,  loans,  age, marital_status, income` 


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import sklearn.model_selection as model_selection
from sklearn.neighbors import KNeighborsClassifier

df = pd.read_csv("/kaggle/input/aimldata/ClassifyRisk.csv")
df
```

### Explorateory Data Analysis 
* Explore the relationship between two varaibles `mortgage, risk`
* Use countplot or crosstab and stacked barplot.
* Findings: higher bad loss for customers with mortgage

```python
sns.countplot(data=df, x="mortgage", hue="risk")
plt.savefig("KNN-03.png")
```

```python
churn_crosstab = pd.crosstab(df["mortgage"], df["risk"],normalize ='index')
churn_crosstab.plot(kind = 'bar', stacked = 'true',grid=True, title = "risk with and without mortage")
plt.xlabel('mortage')
plt.ylabel('Count')
plt.show() 
```

#### Predictors and target variable
* We use $X$ to denote predictors (features):  mortgage,  loans,   age,  marital\_status, income 
* We use $y$ to denote target (class label): risk   

```python
X = df.iloc[:,0:5] 
print(X)
```

```python
y = df.iloc[:,5] 
print(y)
```

### Encode categorical features: ordinal data
* Credit risk data include two categorical features: `Mortgage, marital_status`
* Categorical data must be converted into numerical one before applying KNN
* `mortgage` belongs to ordinal data in the order $n <y$. Use OrdinalEncoder. Use OrdinalEncoder

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories = [['n','y']]) # create an encoder with order
X['mortgage'] = encoder.fit_transform(X[['mortgage']]) # fit encoder with data and transfer data 
X
 
```

### Encode categorical features: nominal data
Since categorical feature `marital_status` has three classes without an order, we use OneHot Encoder
* However, we don't adopt Sklearn's OneHot Encoder because adding columns names is a bit complex 
* Instead, Pandas its own OneHot Encoder method called `pandas.get_dummies`
* A big advantage is that `pandas.get_dummies` manipulates dataframes. 
* New column names are automatically added with a prefix `'marital_status`  and its class labels `married`, `single` or `other`. 

```python
X = pd.get_dummies(X, prefix=['m'], columns=['marital_status'])
X
 
```

### Encode categorical features: nominal data

* Shall we encode the target into numerical value? 
* No need to in Sklearn.  

```python
y = df['risk']
print(y)
```

* But you can encode the categorical target into numerical value using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder() # create an encoder  
y = encoder.fit_transform(df['risk']) # fit encoder with data and transfer data 
print(y[0:10])
```

### Build and train a model
#### Split data
* Split data X, y into training (70%) and test data (30%): `X_train, X_test, y_train, y_test`
* Splitting is at random. Parameter `random_state =7` controls random number seed for repeating the same splitting

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.3, random_state = 7)
```

#### Normalize training data then test data
* We normalize training data `X_Train` to the range $[0,1]$
* Create `MinMaxScaler` and fit it with training data `X_Train`
* In real world, `X_test` is regarded as new data and the scaler is not required to fit with `X_test`  
* We apply the **same scaler** to `X_test`
* **No need to normalize target variable `y` because they represent class labels**

```python
#normalisation
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() 
scaler.fit(X_train)
X_train = scaler.transform(X_train) 
X_test = scaler.transform(X_test)
```

```python
X_train[0:5]
```

```python
X_test[0:5]
```

#### Build and train a model
*   create a KNN model
* train the model on training data

```python
knn = KNeighborsClassifier(n_neighbors=1,weights="uniform", metric="euclidean")
knn.fit(X_train, y_train)
```

### Evaluate the model
* Predict labels of test data
* compare the prediction on the test data between actual labels `y_test` and predicted labels `y_predict` 
    $$ 
    Accuracy  = \frac{\mbox{number of correct predictions}}{\mbox{number of total predictions}}$$
 
* 8 correct predictions in total 10 predictions 
    * 3 actual bad $\to$ predict bad
    * 5 actual good $\to$ predict good
    * 2 actual good $\to$ predict bad
* Accuracy = 8/10=80\% 

```python
from sklearn.metrics import accuracy_score
y_predict = knn.predict(X_test) 
accuracy = accuracy_score(y_test, y_predict)
print(accuracy) 
```

#### Hyperparameters
There are several hyperparameters in KNN.
1. k: the number of neighbour instances. 
2. distances: Euclidean or Manhattan
3. Voting methods: simple unweighted voting or weighted voting.
You need to manually tune these parameters though different combinations and find out which combination is best, i.e., to achieve the best accuracy.  


## Summary
* KNN is a kind of instance learning (supervised learning)
* Instances in k nearest neighbour vote the class of a new data
* Similarity is measured by a distance function
* Instances in neighbour vote the class label: uniform vote and weighted
* Data may need normalisation in kNN
* Categorical data should be converted into numerical ones in distance function
